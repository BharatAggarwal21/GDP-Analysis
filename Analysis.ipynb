# Evaluation

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
np.seterr(divide='ignore', invalid='ignore')

df = pd.read_csv('final_dataset.csv')
countries = df['Country Name'].unique()
col_names = ['Agriculture', 'Education', 'Exports', 'Imports', 'Military', 'RnD', 'Resources', 'Stocks', 'Tax']
```

# Feature selection

## F_regression

```python
from sklearn.feature_selection import f_regression
score_lst = []
variances = []
for i in countries:
    data = df[df['Country Name'] == i][df.columns[2:-1]].values
    target = df[df['Country Name'] == i][['GDP']].values.ravel()
    f_imp, _ = f_regression(data, target)
    f_imp /= np.max(f_imp)
    score_lst.append(f_imp)
df_scores_fr = pd.DataFrame(score_lst, columns=col_names)
df_scores_fr['Country Name'] = countries
df_scores_fr.head()
```

```python
for i in col_names:
    plt.figure(figsize=(20,5))
    plt.bar(df_scores_fr['Country Name'], df_scores_fr[i])
    plt.title(i)
    plt.tick_params(labelrotation = 90)
    plt.show()
```

### PCA

```python
from sklearn.decomposition import PCA
score_lst = []
variances = []
for i in countries:
    data = df[df['Country Name'] == i][df.columns[2:-1]].values
    target = df[df['Country Name'] == i][['GDP']].values.ravel()
    pca = PCA(n_components=1)
    fit = pca.fit(data,target)
    variances.append(fit.explained_variance_ratio_[0])
    score_lst.append(fit.components_[0])
df_scores_pca = pd.DataFrame(score_lst, columns=col_names)
df_scores_pca['Variance'] = variances
df_scores_pca['Country Name'] = countries
df_scores_pca.head()
```

```python
for i in col_names:
    plt.figure(figsize=(20,5))
    plt.bar(df_scores_pca['Country Name'], df_scores_pca[i])
    plt.title(i)
    plt.tick_params(labelrotation = 90)
    plt.show()
```

# Train Models

```python
from sklearn.model_selection import train_test_split
```

## linear regression

```python
from sklearn.linear_model import LinearRegression
accuracies = []
for i in countries:
    data = df[df['Country Name'] == i][df.columns[2:-1]].values
    target = df[df['Country Name'] == i][['GDP']].values.ravel()
    trainx, testx, trainy, testy = train_test_split(data, target, test_size=0.2)
    lr = LinearRegression()
    lr.fit(trainx, trainy)
    accuracies.append(lr.score(testx,testy))
df_lr = pd.DataFrame({'Country Name':countries, 'Accuracy': accuracies})
print ('Mean Accuracy : ', df_lr.Accuracy.values.mean())
df_lr.head()
```

## ridge

```python
from sklearn.linear_model import Ridge
accuracies = []
for i in countries:
    data = df[df['Country Name'] == i][df.columns[2:-1]].values
    target = df[df['Country Name'] == i][['GDP']].values.ravel()
    trainx, testx, trainy, testy = train_test_split(data, target, test_size=0.2)
    ridge = Ridge()
    ridge.fit(trainx, trainy)
    accuracies.append(ridge.score(testx,testy))
df_ridge = pd.DataFrame({'Country Name':countries, 'Accuracy': accuracies})
print ('Mean Accuracy : ', df_ridge.Accuracy.values.mean())
df_ridge.head()
```

## decisiontreeregressor

```python
from sklearn.tree import DecisionTreeRegressor
accuracies = []
for i in countries:
    data = df[df['Country Name'] == i][df.columns[2:-1]].values
    target = df[df['Country Name'] == i][['GDP']].values.ravel()
    trainx, testx, trainy, testy = train_test_split(data, target, test_size=0.2)
    dtr = DecisionTreeRegressor()
    dtr.fit(trainx, trainy)
    accuracies.append(dtr.score(testx,testy))
df_dtr = pd.DataFrame({'Country Name':countries, 'Accuracy': accuracies})
print ('Mean Accuracy : ', df_dtr.Accuracy.values.mean())
df_dtr.head()
```

## randomforestregressor

```python
from sklearn.ensemble import RandomForestRegressor
accuracies = []
score_lst = []
for i in countries:
    data = df[df['Country Name'] == i][df.columns[2:-1]].values
    target = df[df['Country Name'] == i][['GDP']].values.ravel()
    trainx, testx, trainy, testy = train_test_split(data, target, test_size=0.2)
    rfr = RandomForestRegressor(n_estimators=100)
    rfr.fit(trainx, trainy)
    accuracies.append(rfr.score(testx,testy))
    score_lst.append(rfr.feature_importances_)
df_rfr = pd.DataFrame(score_lst, columns=col_names)
df_rfr['Accuracy'] = accuracies
df_rfr['Country Name'] = countries
print ('Mean Accuracy : ', df_rfr.Accuracy.values.mean())
df_rfr.head()
```

```python
plt.figure(figsize=(20,5))
plt.bar(df_rfr['Country Name'], df_rfr['Accuracy'])
plt.title('accuracy of different countries')
plt.tick_params(labelrotation = 90)
plt.show()
```

# Arima

```python
from statsmodels.tsa.arima_model import ARIMA
```

```python
results = []
for i in countries:
    for j in col_names:
        data = df[df['Country Name'] == i][j].values
        y_actual = data[-1]
        data = data[:-1]
        try:
            model = ARIMA(data, order=(1,0,0))
            model_fit = model.fit()
            y_pred = model_fit.forecast()[0][0]
        except:
            y_pred = 0
        rmse = np.sqrt(np.mean((y_pred-y_actual)**2))
        results.append([i,j,y_actual,y_pred,rmse])
df_arima = pd.DataFrame(results, columns = ['Country Name', 'Sector', 'Actual', 'Pred', 'RMSE'])
```

```python
df_arima.head()
```

```python
for i in col_names:
    plt.figure(figsize=(20,5))
    plt.bar(df_arima[df_arima.Sector == i]['Country Name'], df_arima[df_arima.Sector == i]['RMSE'])
    plt.title(i)
    plt.tick_params(labelrotation = 90)
    plt.show()
```

### Country with minimum rmse

```python
mean_rmse = []
for i in countries:
    mean_rmse.append(df_arima[df_arima['Country Name'] == i].RMSE.mean())
mean_rmse = np.array(mean_rmse)
min_rmse_country = countries[mean_rmse.argmin()]
min_rmse_country
```

```python
df_arima[df_arima['Country Name'] == min_rmse_country]
```

```python
results = []
for i in countries:
    data_min_rmse_country = df[df['Country Name'] == i]
    y_actual = data_min_rmse_country.iloc[-1, :].values[-1]
    data_min_rmse_country = data_min_rmse_country.iloc[:-1, :]

    testx = []
    for j in col_names:
        data = data_min_rmse_country[j].values
        try:
            model = ARIMA(data, order=(1,0,0))
            model_fit = model.fit()
            y_pred = model_fit.forecast()[0][0]
        except:
            y_pred = 0
        testx.append(y_pred)
    testx = np.array(testx)
    testx = testx.reshape(1,-1)


    from sklearn.tree import DecisionTreeRegressor
    trainx = df[data_min_rmse_country.columns[2:-1]].values
    trainy = df[['GDP']].values.ravel()
    dtr = DecisionTreeRegressor()
    dtr.fit(trainx, trainy)
    y_pred = dtr.predict(testx)[0]
    error = abs(y_pred-y_actual)/y_actual
    results.append([i, y_actual, y_pred, error])
gdp_pred = pd.DataFrame(results, columns = ['Country Name', 'Actual', 'Predicted', '% Error'])
```

```python
plt.figure(figsize=(20,5))
plt.bar(gdp_pred['Country Name'], gdp_pred['% Error'])
plt.title('Error % in GDP prediction with predicted values')
plt.tick_params(labelrotation = 90)
plt.show()
```
